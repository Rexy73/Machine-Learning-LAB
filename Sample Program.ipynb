```python
#SAMPLE PROGRAM
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
```

```python
iris=load_iris()
x=iris.data
y=iris.target
print("Keys of iris datasets \n{}".format(iris.keys()))
print("Values of iris datasets \n{}".format(iris.values()))
print(iris['target_names'])
print("Feature names:\n{}".format(iris['feature_names']))
```

```python
x.shape
```

```python
y.shape
```

```python
import pandas as pd
iris_df=pd.DataFrame(x,columns=iris.feature_names)
iris_df['species']=pd.Categorical.from_codes(y,iris.target_names)
print(iris_df)
iris_df.head(5)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
DTC=DecisionTreeClassifier()
DTC.fit(x_train,y_train)
```

```python
predictions=DTC.predict(x_test)
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,predictions))
```

## Simple Imputer

```python
import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer
import warnings
warnings.filterwarnings('ignore')
data = {
    'Name': ['ME', 'Adil', np.nan, 'Yathin', 'Navneet', 'Carli', 'Steve'],
    'Age': [21, 23, np.nan, 19, 25, np.nan, 15],
    'Sex': ['M', np.nan, np.nan, 'M', 'F', 'F', 'M'],
    'Goals': [5, 10, np.nan, 19, 5, 0, 7],
    'Assists': [7, 4, np.nan, 9, 3, 6, 4],
    'Value': [55, 84, np.nan, 90, 63, 15, 46]
}

df = pd.DataFrame(data=data)
print("Original Data:")
print(df.head(n=7))
print("\n")
imputer = SimpleImputer(strategy='mean')
df['Age'] = imputer.fit_transform(df[['Age']])
print("Data after Imputation (Age):")
print(df.head(n=7))
drop_rows = df.dropna(how='all')
print("\nRows dropped (where all values are NaN):")
print(drop_rows)

```

## Label Encoding

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder

cancer_data = pd.read_csv("breast-cancer.csv")

print(cancer_data.sample(5))
print(cancer_data['diagnosis'].value_counts())
label_encode = LabelEncoder()

labels = label_encode.fit_transform(cancer_data['diagnosis'])
cancer_data['target'] = labels
print(cancer_data.sample(5))
print(cancer_data['target'].value_counts())


iris_data = pd.read_csv("Iris.csv")
print(iris_data.head())
print(iris_data['Species'].value_counts())
label_encode = LabelEncoder()

iris_labels = label_encode.fit_transform(iris_data['Species'])
iris_data['target'] = iris_labels
print(iris_data['target'].value_counts())
```

## Normalization and Standardization

```python
from sklearn.datasets import load_iris
from sklearn.preprocessing import MinMaxScaler
import pandas as pd
iris = load_iris()
x = pd.DataFrame(iris.data, columns=iris.feature_names)
print("Data before Normalization:")
print(x.head())
scaler = MinMaxScaler()
data_normalized = scaler.fit_transform(x)
normalized_data_df = pd.DataFrame(data_normalized, columns=iris.feature_names)
print("\nData after normalization:")
print(normalized_data_df.head())
```

```python
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
import pandas as pd
iris = load_iris()
x = pd.DataFrame(iris.data, columns=iris.feature_names)
print("\nData before Standardization:")
print(x.head())
scaler = StandardScaler()
data_standardized = scaler.fit_transform(x)
standardized_data_df = pd.DataFrame(data_standardized, columns=iris.feature_names)
print("\nData after standardization:")
print(standardized_data_df.head())
```

```python
import numpy as np
import pandas as pd
from sklearn.feature_selection import SelectKBest, f_classif
from sklearn.preprocessing import StandardScaler

# Populating with random data to match the length of 'Study Hours'
data = pd.DataFrame({
    'Student Name': [f'Student{i}' for i in range(1, 11)],
    'Study Hours': [2, 3, 5, 6, 8, 9, 4, 7, 5, 6],
    'Sleep Hours': [6, 7, 8, 5, 7, 6, 8, 7, 6, 8],
    'Attendance': [75, 80, 95, 85, 90, 100, 70, 88, 82, 92],
    'Marks': [55, 60, 75, 82, 88, 95, 65, 85, 78, 90],
    'TestScores': [70, 75, 85, 80, 90, 95, 60, 85, 75, 88],
    'ProjectScores': [75, 80, 88, 85, 92, 98, 65, 80, 82, 90]
})

# Create target variable
data['Pass'] = (data['TestScores'] + data['ProjectScores'] > 150).astype(int)

# Dropping 'Pass' (target) and 'Student Name' (string) so scaling works
X = data.drop(['Pass', 'Student Name'], axis=1) 
y = data['Pass']

print("--- Original Features (First 5 rows) ---")
print(X.head())
print("\n")

# Scale the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Converting the scaled numpy array back to a DataFrame for cleaner printing
X_scaled_df = pd.DataFrame(X_scaled, columns=X.columns)

print("--- Scaled Features (First 5 rows) ---")
print(X_scaled_df.head())
```
